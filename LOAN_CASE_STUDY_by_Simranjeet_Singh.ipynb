<p><b>Problem:</b> Identification of patterns using EDA which indicate if a client has difficulty paying their installments which may be used for taking actions such as denying the loan, reducing the amount of loan, lending (to risky applicants) at a higher interest rate, etc. This will ensure that the consumers capable of repaying the loan are not rejected.</p>

<p><b>Approach :</b>Will be Analyzing the variables and then cleaning them by dealing with  the outliers and null values if any.Univariate, bivariate and multivariate analysis will be performed accordingly, to find out which variables/factors have effect on a person defaulting a loan or not.</p>

```python
columns=list(data_new.columns)
list1=["SK_ID_PREV", "SK_ID_CURR","AMT_ANNUITY_x","AMT_APPLICATION","AMT_CREDIT_x","AMT_DOWN_PAYMENT","RATE_DOWN_PAYMENT","RATE_INTEREST_PRIMARY","RATE_INTEREST_PRIVILEGED","CNT_PAYMENT","AMT_INCOME_TOTAL","AMT_CREDIT_y","AMT_ANNUITY_y","AMT_GOODS_PRICE","CNT_FAM_MEMBERS","EXT_SOURCE_1","EXT_SOURCE_2","EXT_SOURCE_3","Age","TARGET"]
for ele in list1:
    columns.remove(ele)
for col in columns:
    if data_new[col].dtype=="int64":
        q3, q1 = np.percentile(data_new[col], [75 ,25])
        iqr = q3 - q1
        median_col=data_new[col].median()
        data_new[col] = np.where(data_new[col] >(median_col+(1.5*iqr)),median_col, data_new[col])
        data_new[col] = np.where(data_new[col] <(median_col-(1.5*iqr)) ,median_col, data_new[col])
    elif data_new[col].dtype=="float64":
        q3, q1 = np.percentile(data_new[col], [75 ,25])
        iqr = q3 - q1
        median_col=data_new[col].median()
        data_new[col] = np.where(data_new[col] >(median_col+(1.5*iqr)),median_col, data_new[col])
        data_new[col] = np.where(data_new[col] <(median_col-(1.5*iqr)),median_col, data_new[col])
```
